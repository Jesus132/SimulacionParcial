# The Algoritmo Wavefront

Before we start, we import some preliminary libraries. We will also import (below) the accompanying `lorenz.py` file, which contains the actual solver and plotting routine.

```python
import random
m=int(input("Digita la cantidad de filas: "))
n=int(input("Digita la cantidad de columnas: "))
fila,col=[],[]
valor,coorF,coorC=[],[],[]

def Recorrido_C(matriz, filas_final, colu_final):
  i=0
  while True:
    menor=[]
    F=[]
    C=[]
    if(matriz[filas_final + 1][colu_final] != "#" and matriz[filas_final + 1][colu_final] != "-" and matriz[filas_final + 1][colu_final] != "x"):
      menor.append(int(matriz[filas_final + 1][colu_final]))
      F.append(filas_final + 1)
      C.append(colu_final)
    if(matriz[filas_final - 1][colu_final] != "#" and matriz[filas_final - 1][colu_final] != "-") and matriz[filas_final - 1][colu_final] != "x":
      menor.append(int(matriz[filas_final - 1][colu_final]))
      F.append(filas_final - 1)
      C.append(colu_final)
    if(matriz[filas_final][colu_final + 1] != "#" and matriz[filas_final][colu_final + 1] != "-") and matriz[filas_final][colu_final + 1] != "x":
      menor.append(int(matriz[filas_final][colu_final + 1]))
      F.append(filas_final)
      C.append(colu_final + 1)
    if(matriz[filas_final][colu_final - 1] != "#" and matriz[filas_final][colu_final - 1] != "-") and matriz[filas_final][colu_final - 1] != "x":
      menor.append(int(matriz[filas_final][colu_final - 1]))
      F.append(filas_final)
      C.append(colu_final - 1)
    for men in range(len(menor)):
      if(int(menor[men]) == min(menor)):
        valor.append(0)
        coorF.append(0)
        coorC.append(0)
        valor[i] = menor[men]
        coorF[i] = F[men]
        coorC[i] = C[men]
    filas_final=coorF[i]
    colu_final=coorC[i]
    if(valor[i] != 0):
      print("Valor de la casilla: ", valor[i])
      print("Coordenadas x: ", coorF[i], " y: ",coorC[i])
    if(valor[i] == 1):
      break
    i+=1

def Recorrido_O(matriz, filas_final, colu_final):
  i=0
  while True:
    menor=[]
    F=[]
    C=[]
    if(matriz[filas_final + 1][colu_final] != "#" and matriz[filas_final + 1][colu_final] != "-" and matriz[filas_final + 1][colu_final] != "x"):
      menor.append(int(matriz[filas_final + 1][colu_final]))
      F.append(filas_final + 1)
      C.append(colu_final)
    if(matriz[filas_final - 1][colu_final] != "#" and matriz[filas_final - 1][colu_final] != "-") and matriz[filas_final - 1][colu_final] != "x":
      menor.append(int(matriz[filas_final - 1][colu_final]))
      F.append(filas_final - 1)
      C.append(colu_final)
    if(matriz[filas_final][colu_final + 1] != "#" and matriz[filas_final][colu_final + 1] != "-") and matriz[filas_final][colu_final + 1] != "x":
      menor.append(int(matriz[filas_final][colu_final + 1]))
      F.append(filas_final)
      C.append(colu_final + 1)
    if(matriz[filas_final][colu_final - 1] != "#" and matriz[filas_final][colu_final - 1] != "-") and matriz[filas_final][colu_final - 1] != "x":
      menor.append(int(matriz[filas_final][colu_final - 1]))
      F.append(filas_final)
      C.append(colu_final - 1)

    if(matriz[filas_final- 1][colu_final - 1] != "#" and matriz[filas_final - 1][colu_final - 1] != "-") and matriz[filas_final - 1][colu_final - 1] != "x":
      menor.append(int(matriz[filas_final - 1][colu_final - 1]))
      F.append(filas_final - 1)
      C.append(colu_final - 1)

    if(matriz[filas_final + 1][colu_final + 1] != "#" and matriz[filas_final + 1][colu_final + 1] != "-") and matriz[filas_final + 1][colu_final + 1] != "x":
      menor.append(int(matriz[filas_final + 1][colu_final + 1]))
      F.append(filas_final + 1)
      C.append(colu_final + 1)
    
    if(matriz[filas_final + 1][colu_final - 1] != "#" and matriz[filas_final + 1][colu_final - 1] != "-") and matriz[filas_final + 1][colu_final - 1] != "x":
      menor.append(int(matriz[filas_final + 1][colu_final - 1]))
      F.append(filas_final + 1)
      C.append(colu_final - 1)
    
    if(matriz[filas_final - 1][colu_final + 1] != "#" and matriz[filas_final - 1][colu_final + 1] != "-") and matriz[filas_final - 1][colu_final + 1] != "x":
      menor.append(int(matriz[filas_final - 1][colu_final + 1]))
      F.append(filas_final - 1)
      C.append(colu_final + 1)
    
    for men in range(len(menor)):
      if(int(menor[men]) == min(menor)):
        valor.append(0)
        coorF.append(0)
        coorC.append(0)
        valor[i] = menor[men]
        coorF[i] = F[men]
        coorC[i] = C[men]
    filas_final=coorF[i]
    colu_final=coorC[i]
    if(valor[i] != 0):
      print("Valor de la casilla: ", valor[i])
      print("Coordenadas x: ", coorF[i], " y: ",coorC[i])
    if(valor[i] == 1):
      break
    i+=1

def Recorrido_Cuatro(matriz):
  i=0
  while True:
    if(len(fila) == i):
      break
    if(fila[i] + 1 < m and col[i] + 1 < n):
    #Arriba
      if(matriz[fila[i] + 1][col[i]] == "#" and fila[i] + 1 >= 0):
        matriz[fila[i] + 1][col[i]]=str(int(matriz[fila[i]][col[i]]) + 1)
        fila.append(fila[i] + 1)
        col.append(col[i])
    #Abajo
      if(matriz[fila[i] - 1][col[i]] == "#" and fila[i] - 1 >= 0):
        matriz[fila[i] - 1][col[i]]=str(int(matriz[fila[i]][col[i]]) + 1)
        fila.append(fila[i] - 1)
        col.append(col[i])
    #Izquierda
      if(matriz[fila[i]][col[i] - 1] == "#" and col[i] - 1 >= 0):
        matriz[fila[i]][col[i] - 1]=str(int(matriz[fila[i]][col[i]]) + 1)
        fila.append(fila[i])
        col.append(col[i] - 1)
    #Derecha
      if(matriz[fila[i]][col[i] + 1] == "#" and col[i] + 1 >= 0):
        matriz[fila[i]][col[i] + 1]=str(int(matriz[fila[i]][col[i]]) + 1)
        fila.append(fila[i])
        col.append(col[i] + 1)
    i+=1

def Recorrido_Ocho(matriz):
  i=0
  while True:
    if(len(fila) == i):
      break
    if(fila[i] + 1 < m and col[i] + 1 < n):
    #Arriba
      if(matriz[fila[i] + 1][col[i]] == "#" and fila[i] + 1 >= 0):
        matriz[fila[i] + 1][col[i]]=str(int(matriz[fila[i]][col[i]]) + 1)
        fila.append(fila[i] + 1)
        col.append(col[i])
    #Abajo
      if(matriz[fila[i] - 1][col[i]] == "#" and fila[i] - 1 >= 0):
        matriz[fila[i] - 1][col[i]]=str(int(matriz[fila[i]][col[i]]) + 1)
        fila.append(fila[i] - 1)
        col.append(col[i])
    #Izquierda
      if(matriz[fila[i]][col[i] - 1] == "#" and col[i] - 1 >= 0):
        matriz[fila[i]][col[i] - 1]=str(int(matriz[fila[i]][col[i]]) + 1)
        fila.append(fila[i])
        col.append(col[i] - 1)
    #Derecha
      if(matriz[fila[i]][col[i] + 1] == "#" and col[i] + 1 >= 0):
        matriz[fila[i]][col[i] + 1]=str(int(matriz[fila[i]][col[i]]) + 1)
        fila.append(fila[i])
        col.append(col[i] + 1)
    #Arriba-Derecha
      if(matriz[fila[i] + 1][col[i] + 1] == "#" and fila[i] + 1 >= 0 and col[i] + 1 >= 0):
        matriz[fila[i] + 1][col[i] + 1]=str(int(matriz[fila[i]][col[i]]) + 1)
        fila.append(fila[i] + 1)
        col.append(col[i] + 1)
    #Arriba-Izquierda
      if(matriz[fila[i] + 1][col[i] - 1] == "#" and fila[i] + 1 >= 0 and col[i] - 1 >= 0):
        matriz[fila[i] + 1][col[i] - 1]=str(int(matriz[fila[i]][col[i]]) + 1)
        fila.append(fila[i] + 1)
        col.append(col[i] - 1)
    #Abajo-Derecha
      if(matriz[fila[i] - 1][col[i] + 1] == "#" and fila[i] - 1 >= 0 and col[i] + 1 >= 0):
        matriz[fila[i] - 1][col[i] + 1]=str(int(matriz[fila[i]][col[i]]) + 1)
        fila.append(fila[i] - 1)
        col.append(col[i] + 1)
    #Abajo-Izquierda
      if(matriz[fila[i] - 1][col[i] - 1] == "#" and fila[i] - 1 >= 0 and col[i] - 1 >= 0):
        matriz[fila[i] - 1][col[i] - 1]=str(int(matriz[fila[i]][col[i]]) + 1)
        fila.append(fila[i] - 1)
        col.append(col[i] - 1)
    i+=1

if __name__ == '__main__': 
  matriz=[["#" for i in range(n)]for j in range(m)]
  fila_final=int(input("Digita la de fila final: "))
  col_final=int(input("Digita la de col final: "))
  fila_inicio=int(input("Digita la de fila inicio: "))
  col_inicio=int(input("Digita la de col inicio: "))
  matriz[fila_inicio][col_inicio]="x"
  matriz[fila_final][col_final]="1"
  
  o=random.randrange(m*n)
  for i in range(o):
    fila_o=random.randrange(m)
    col_o=random.randrange(n)
    if(fila_o & col_o != fila_inicio & col_inicio and fila_o & col_o != fila_final & col_final):
      matriz[fila_o][col_o]="-"

  for i in matriz:
    print(i)
  
  Metodo=int(input("Digita 0 para ir a 4Celdas o 1 para ir a 8Celdas: "))
  fila.append(fila_final)
  col.append(col_final)
  if(Metodo == 0):
    Recorrido_Cuatro(matriz)
    print("-------------------------------------")
    for i in matriz:
      print(i)
    Recorrido_C(matriz, fila_inicio, col_inicio)
  if(Metodo == 1):
    Recorrido_Ocho(matriz)
    print("-------------------------------------")
    for i in matriz:
      print(i)
    Recorrido_O(matriz, fila_inicio, col_inicio)
```
